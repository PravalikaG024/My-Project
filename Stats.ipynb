```python
import numpy as np
import pandas as pd
import statistics
import scipy
import math
from scipy.stats import skew, kurtosis
```

```python
x = [8.0, 1, 2.5, 4, 28.0]
mean = np.mean(x)   
print (mean)
```

```python
x = [8.0, 1, 2.5, 4, 28.0]
mean = statistics.mean(x)
variance = statistics.variance(x,mean)
standard_deviation = variance ** 0.5
print (standard_deviation) 
```

```python

x = [8.0, 1, 2.5, 4, 28.0]
standard_deviation = statistics.stdev(x) 
print (standard_deviation)
```

```python
data = np.array([2, 4, 8, 10, 12, 67, 23, 32,14,10,54])

# 25th percentile (same as first quartile)
quantile_25 = np.quantile(data, 0.25)
quantile_50 = np.quantile(data, 0.50)
quantile_75 = np.quantile(data,0.75)
```

```python

Q3 = np.quantile(data,0.75)
Q1 = np.quantile(data,0.25) 
IQR = Q3 - Q1 
print(IQR)
```

```python
lb = Q1 - 1.5*IQR
ub = Q3 + 1.5*IQR
print(lb)
print(ub)
```

```python
x = [8.0, 1, 2.5, 4, 28.0]
median = np.median(x)
print (median)
```

```python
z = [2, 3, 2, 8, 12]
mode= statistics.mode(z) 
print (mode)
```
