```python
#Write a Python script to sort (ascending and descending) a dictionary by value.
# Create a dictionary
marks = {'John': 85, 'Emma': 92, 'Bob': 75, 'Lisa': 90}

# Sort by value (small to big)
#Here Key argument is taken into consideration as it to compare with x itself using lamda function
asc = sorted(marks.items(), key=lambda x: x[1])
print("Ascending:", asc)

# Sort by value (big to small)
desc = sorted(marks.items(), key=lambda x: x[1], reverse=True)
print("Descending:", desc)
```

```python
#Write a Python script to add a key to a dictionary.

# Create a dictionary 'd' with two key-value pairs.
d = {0: 10, 1: 20}
print(d)

# Update the dictionary 'd' by adding a new key-value pair {2: 30}.
d.update({2: 30})
print(d) 
```

```python
#Write a Python script to concatenate the following dictionaries to create a new one.

dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}

# Create an empty dictionary 'dic4' that will store the combined key-value pairs from 'dic1', 'dic2', and 'dic3'.
dic4 = {}

# Iterate through each dictionary ('dic1', 'dic2', and 'dic3') using a loop.
for d in (dic1, dic2, dic3):
    # Update 'dic4' by adding the key-value pairs from the current dictionary 'd'.
    dic4.update(d)
print(dic4) 

```

```python
#Write a Python script to check whether a given key already exists in a dictionary.

d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

def is_key_present(x):
    if x in d:
        print('Key is present in the dictionary')
    else:
        print('Key is not present in the dictionary')


is_key_present(5)
is_key_present(9) 
```

```python
#Iterate over dictionaries using for loops
#Write a Python program to iterate over dictionaries using for loops.
d = {'x': 10, 'y': 20, 'z': 30} 
for dict_key, dict_value in d.items():
    # Print the key followed by '->' and the corresponding value.
    print(dict_key, '->', dict_value)	
```

```python
#Write a Python program to sum all the items in a dictionary.
my_dict = {'data1': 100, 'data2': -54, 'data3': 247}
result = sum(my_dict.values())
# Print the result, which is the sum of the values.
print(result) 
```

```python
# Write a Python program to multiply all the items in a dictionary.
my_dict = {'data1': 100, 'data2': -54, 'data3': 247}
result = 1
for key in my_dict:
    result = result * my_dict[key]
print(result) 
```
